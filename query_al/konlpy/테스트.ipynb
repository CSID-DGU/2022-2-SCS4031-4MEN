```python
from collections import Counter

from konlpy.corpus import kolaw
from konlpy.tag import Hannanum
from konlpy.utils import concordance, pprint
from matplotlib import pyplot

# c = kolaw.open('constitution.txt').read()
# 기본 내장된 대한민국 헌법 내용

from konlpy.tag import *
hannanum = Hannanum()
kkma = Kkma()
okt = Okt()
# 명사: nouns
# 형태소: morphs
# 품사부착: pos

# APSAC, NCAC 면담기법은 아동과의 피해 면담 기법이지만 '학대 피해 장애인에 대한 수사 전 면담기법에 관한 연구'
# 에서는 장애인 면담 기법으로 참고하고 있다. 
# 면담 개시 시에 장애인의 경우 잘 모르는 사람과의 면담 시 거부감, 불안을 느낄 수 있다
# 장애의 정도, 나이 등 상황에 따라 대화 기법을 탄력적으로 바꿔야 한다.  
# 호칭에 대한 확인 : 연령과 장애수준에 맞추어진 호칭을 정해야하고 애매할 경우 의사를 물어봐야한다.
# 이해의 확인 : 면담 대상이 이해를 확실히 했는지 재차 묻고 못했을 경우 질문을 바꾼다. 또 지적, 발달 장애의 경우 대답이 오래 걸리므로 이를 고려
# 이를 라포(rapport) 형성이라고 한다. 
```

```python
# 위의 hanspell 보다 더 나은 성능과 기능을 보인다.
from symspellpy_ko import KoSymSpell, Verbosity

sym_spell = KoSymSpell()
sym_spell.load_korean_dictionary(decompose_korean=True, load_bigrams=True)
```

### 유사한 어절 찾기

```python
# 유사한 어절 찾기
term = "주세오"
for suggestion in sym_spell.lookup(term, Verbosity.ALL):
  print(suggestion.term, suggestion.distance, suggestion.count)
```

```python
term = "때리다"
for suggestion in sym_spell.lookup(term, Verbosity.ALL):
  print(suggestion.term, suggestion.distance, suggestion.count)
```

```python
# 한스펠을 사용해 고친 단어
text = "저부 말헤 주세오"
for suggestion in sym_spell.lookup_compound(text, max_edit_distance=2):
  print(suggestion.term, suggestion.distance, suggestion.count)
```
